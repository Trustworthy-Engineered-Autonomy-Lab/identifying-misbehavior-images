```python
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Ensure reproducibility
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

# Directories for normal and misbehavior images
normal_dir = '/home/UFAD/mohitkukreja/Documents/data_perfect_driving/images'
blurry_dir = '/home/UFAD/mohitkukreja/Documents/data_perfect_driving/images_blurred'
dirt_dir = '/home/UFAD/mohitkukreja/Documents/data_perfect_driving/images_dirt/'
tape_dir = '/home/UFAD/mohitkukreja/Documents/data_perfect_driving/images_tape/'
```

```python
# Data Loading and Preprocessing
def load_and_preprocess_images(folder_path, num_images, target_size=(64, 64)):
    """Load and preprocess specified number of images from a folder"""
    images = []
    count = 0
    for filename in os.listdir(folder_path):
        if count >= num_images:
            break
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(folder_path, filename)
            img = load_img(img_path, target_size=target_size, color_mode='grayscale')
            img_array = img_to_array(img)
            img_array = img_array / 255.0
            images.append(img_array)
            count += 1
    return np.array(images)

# Build Autoencoder
def build_autoencoder(input_shape):
    """Build a simple autoencoder model"""
    # Encoder
    input_layer = layers.Input(shape=input_shape)
    
    # Encoding layers
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_layer)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    
    # Decoding layers
    x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)
    
    # Create model
    autoencoder = models.Model(input_layer, x)
    autoencoder.compile(optimizer='adam', loss='mse')
    
    return autoencoder

def calculate_reconstruction_error(original, reconstructed):
    """Calculate MSE reconstruction error"""
    return np.mean(np.square(original - reconstructed), axis=(1,2,3))

def plot_reconstruction_errors(errors_dict, threshold):
    """Plot reconstruction errors distribution"""
    plt.figure(figsize=(15, 6))
    for category, errors in errors_dict.items():
        sns.kdeplot(errors, label=category)
    plt.axvline(x=threshold, color='r', linestyle='--', label='Threshold')
    plt.xlabel('Reconstruction Error')
    plt.ylabel('Density')
    plt.title('Distribution of Reconstruction Errors by Category')
    plt.legend()
    plt.show()

def plot_sample_results(original_images, reconstructed_images, errors, threshold, title):
    """Plot sample correct and incorrect predictions"""
    is_anomaly = errors > threshold
    fig, axes = plt.subplots(4, 4, figsize=(15, 15))
    fig.suptitle(title)
    
    # Plot 2 correct and 2 incorrect predictions for both normal and anomaly cases
    correct_normal = np.where((~is_anomaly))[0][:2]
    incorrect_normal = np.where(is_anomaly)[0][:2]
    
    # Plot samples
    for idx, ax_row in enumerate(axes):
        if idx < 2:  # Correct predictions
            for j, ax in enumerate(ax_row[:2]):
                if len(correct_normal) > j:
                    ax.imshow(original_images[correct_normal[j]].squeeze(), cmap='gray')
                    ax.set_title(f'Correct\nError: {errors[correct_normal[j]]:.4f}')
            for j, ax in enumerate(ax_row[2:]):
                if len(correct_normal) > j:
                    ax.imshow(reconstructed_images[correct_normal[j]].squeeze(), cmap='gray')
                    ax.set_title('Reconstruction')
        else:  # Incorrect predictions
            for j, ax in enumerate(ax_row[:2]):
                if len(incorrect_normal) > j:
                    ax.imshow(original_images[incorrect_normal[j]].squeeze(), cmap='gray')
                    ax.set_title(f'Incorrect\nError: {errors[incorrect_normal[j]]:.4f}')
            for j, ax in enumerate(ax_row[2:]):
                if len(incorrect_normal) > j:
                    ax.imshow(reconstructed_images[incorrect_normal[j]].squeeze(), cmap='gray')
                    ax.set_title('Reconstruction')
        
    for ax in axes.flat:
        ax.axis('off')
    plt.tight_layout()
    plt.show()

# Main execution
def main():
    # Load data
    print("Loading data...")
    normal_images = load_and_preprocess_images(normal_dir, 10000)
    blurry_images = load_and_preprocess_images(blurry_dir, 500)
    taped_images = load_and_preprocess_images(tape_dir, 500)
    dirty_images = load_and_preprocess_images(dirt_dir, 500)
    
    # Split normal images into train and validation
    train_size = int(0.8 * len(normal_images))
    train_normal = normal_images[:train_size]
    val_normal = normal_images[train_size:]
    
    # Build and train autoencoder
    print("Training autoencoder...")
    autoencoder = build_autoencoder(normal_images[0].shape)
    history = autoencoder.fit(
        train_normal, train_normal,
        epochs=50,
        batch_size=32,
        validation_data=(val_normal, val_normal),
        verbose=1
    )
    
    # Calculate reconstruction errors for each category
    print("Calculating reconstruction errors...")
    reconstructed_normal = autoencoder.predict(normal_images)
    reconstructed_blurry = autoencoder.predict(blurry_images)
    reconstructed_taped = autoencoder.predict(taped_images)
    reconstructed_dirty = autoencoder.predict(dirty_images)
    
    errors_dict = {
        'Normal': calculate_reconstruction_error(normal_images, reconstructed_normal),
        'Blurry': calculate_reconstruction_error(blurry_images, reconstructed_blurry),
        'Taped': calculate_reconstruction_error(taped_images, reconstructed_taped),
        'Dirty': calculate_reconstruction_error(dirty_images, reconstructed_dirty)
    }
    
    # Calculate threshold (95th percentile of normal errors)
    threshold = np.percentile(errors_dict['Normal'], 95)
    
    # Plot reconstruction error distributions
    plot_reconstruction_errors(errors_dict, threshold)
    
    # Evaluate each category
    categories = {'Normal': normal_images, 'Blurry': blurry_images, 
                 'Taped': taped_images, 'Dirty': dirty_images}
    
    print("\nEvaluation Results:")
    print("-" * 50)
    
    for category, images in categories.items():
        reconstructed = autoencoder.predict(images)
        errors = calculate_reconstruction_error(images, reconstructed)
        predictions = errors > threshold
        
        true_labels = np.zeros(len(predictions)) if category == 'Normal' else np.ones(len(predictions))
        
        # Print category-specific metrics
        print(f"\nCategory: {category}")
        print(classification_report(true_labels, predictions))
        
        # Plot sample results
        plot_sample_results(images, reconstructed, errors, threshold, 
                          f'Sample Results for {category} Images')
        
        # Print specific numbers
        if category != 'Normal':
            incorrect = np.sum(~predictions)
            print(f"Number of {category} images misclassified as normal: {incorrect}")
            print(f"Percentage: {(incorrect/len(predictions))*100:.2f}%")
    
    # Plot training history
    plt.figure(figsize=(10, 4))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Autoencoder Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
